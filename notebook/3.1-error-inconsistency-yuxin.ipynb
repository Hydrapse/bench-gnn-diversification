```python
import matplotlib.pyplot as plt
import numpy as np
import torch
import os

# === Configuration ===
selected_datasets = ["arxiv", "squirrel", "actor", "wikics"]
depth_file = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep_depth/logit/trend_value.pt"
adj_file = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep_adj/logit/values.pt"
jk_res_file = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep_jk_res/logit/values.pt"


# === Load Data ===
depth_grids = torch.load(depth_file)
adj_grids = torch.load(adj_file)
jk_res_grids = torch.load(jk_res_file)

# === Labels ===
depth_labels = [f"L={i}" for i in range(7)]
adj_labels_raw = ["I", "sym", "rw", "high", "attn", "second", "bern"]
adj_labels_latex = [r"$I$", r"$A_{\mathrm{sym}}$", r"$A_{\mathrm{rw}}$",
                    r"$L_{\mathrm{sym}}$", r"$A_{\mathrm{attn}}$", r"$A_{\mathrm{sym}}^2$", r"$A_{\mathrm{bern}}$"]
jk_res_labels = ["none", "jk-cat", "jk-max", "jk-attn", "res-sum", "res-cat", "res-incep"]

# === Figure Setup ===
n_rows, n_cols = len(selected_datasets), 3
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(10, 1.8 * n_rows))
fig.subplots_adjust(hspace=0.5, wspace=0.3)

for row_idx, dataset in enumerate(selected_datasets):
    depth_map = depth_grids.get(dataset, {}).get("GCN", None)
    adj_map = adj_grids.get(dataset, {}).get("GCN", None)
    jk_res_map = jk_res_grids.get(dataset, {}).get("GCN", None)
    heatmaps = [depth_map, adj_map, jk_res_map]

    for col_idx, heatmap in enumerate(heatmaps):
        ax = axes[row_idx, col_idx]
        if heatmap is None:
            ax.axis("off")
            continue

        im = ax.imshow(heatmap, cmap='viridis', vmin=0)
        ax.set_aspect('auto')  #  Add this line here

        # Titles
        if row_idx == 0:
            if col_idx == 0: ax.set_title("Depths")
            elif col_idx == 1: ax.set_title("Graph Filters")
            elif col_idx == 2: ax.set_title("Skip Connections")

        if col_idx == 0:
            ax.set_ylabel(dataset, fontsize=12)
            ax.set_xticks(np.arange(7))
            ax.set_yticks(np.arange(7))
            ax.set_xticklabels(depth_labels, fontsize=8)
            ax.set_yticklabels(depth_labels, fontsize=8)
        elif col_idx == 1:
            ax.set_xticks(np.arange(len(adj_labels_latex)))
            ax.set_xticklabels(adj_labels_latex, rotation=45, ha='right')
            ax.set_yticks(np.arange(len(adj_labels_latex)))
            ax.set_yticklabels(adj_labels_latex)
        elif col_idx == 2:
            ax.set_xticks(np.arange(len(jk_res_labels)))
            ax.set_xticklabels(jk_res_labels, rotation=45, ha='right')
            ax.set_yticks(np.arange(len(jk_res_labels)))
            ax.set_yticklabels(jk_res_labels)

    # Colorbar aligned to the middle of the row
    cbar_ax = fig.add_axes([0.93, 0.8 - row_idx * 0.235, 0.015, 0.15])
    plt.colorbar(im, cax=cbar_ax)

plt.tight_layout(rect=[0, 0, 0.91, 1])
save_path = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep/figs/combined_heatmap_rectangular.png"
plt.savefig(save_path)
plt.show()
# plt.close()
# print(f" Saved updated heatmap figure to: {save_path}")
```
