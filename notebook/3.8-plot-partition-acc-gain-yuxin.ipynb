```python
import matplotlib.pyplot as plt
import numpy as np

# Generate 18 random values (simulating the input)
np.random.seed(0)
relative_accuracy_improvements = [
    [-0.3891797482, -0.3006331895, -0.297638077, -0.2806822697, -0.2633427777, 0.1516174525, -0.02794687527, 0.1802577593, 0.5859017683],
    [-0.04998601477, -0.006199673214, -0.02271440771, -0.02719677045, -0.01187219619,  0.297975382,0.09124111197, 0.3055273793, 0.6964817168]
]


values = np.array(relative_accuracy_improvements).reshape(2, 9)/ 10 # 10 datasets

# Labels
x_labels = [
    "Random", "Pagerank", "Cluster-coef",
    "Max Neigh. Label Ratio", "Neigh. Label Entropy","Intra-class Degree", 
    "Intra-class Feat Sim.",  "Intra-class Label Sim.","Homophily"
]
group_labels = [
    "Random", "Topology-based Metrics", "Neighborhood Label Informativeness", "Intra-class Graph Metrics"
]

# Plot
x = np.arange(len(x_labels))
width = 0.35

fig, ax = plt.subplots(figsize=(6, 4.2))
bar1 = ax.bar(x - width/2, values[0], width, label='Train on domain')
bar2 = ax.bar(x + width/2, values[1], width, label='Finetune on domain')

# Axes and labels
ax.set_ylabel('Relative Accuracy Improvements')
ax.set_title('Average Accuracy Diff on Domains')
ax.set_xticks(x)
ax.set_xticklabels(x_labels, rotation=45, ha='right')
ax.axhline(0, color='black', linewidth=0.8)
ax.legend(loc='upper left')
ax.set_yticklabels([f"{int(tick * 100)}%" for tick in ax.get_yticks()])


# Vertical lines for group separation
vlines = [0.5, 2.5, 4.5]
for v in vlines:
    ax.axvline(x=v, color='lightgray', linestyle='--', linewidth=1)

group_labels = {
    0: "Random",
    1.5: "Topology-based \nMetrics",
    3.5: "Neighborhood \nLabel \nInformativeness",
    6.5: "Intra-class Graph Metrics"
}

for xpos, label in group_labels.items():
    ax.text(xpos, -0.044, label, fontsize=6.5, ha='center', va='bottom', rotation=0, color='gray')


plt.tight_layout()
plt.savefig('relative_accuracy_improvements.png', dpi=300, bbox_inches='tight')
plt.show()
```
