```python
import os
from omegaconf import OmegaConf
import re

# Directory containing the YAML config files
# config_dir = "/home/yuxin/Projects/gnn-moe/conf/exp_sota"
config_dir = "../conf/exp_sota"

# Parameters you care about
keys_to_extract = ["model.dropout", "train.epoch", "model.hidden_dim"]

# Result dictionary
standard_config = {}

# Loop over all YAML files
for file in os.listdir(config_dir):
    if not file.endswith(".yaml"):
        continue

    # e.g., arxiv_gcn.yaml → dataset='arxiv', model='gcn'
    dataset_model = file.replace(".yaml", "")
    parts = dataset_model.split("_")
    if len(parts) == 3:
        # Handle cases like 'amazon_ratings_gcn.yaml'
        dataset = "_".join(parts[:-1])
        model = parts[-1]
        if model not in models:
            continue
    else:
        # Handle cases like 'arxiv_gcn.yaml'
        if len(parts) != 2:
            continue
        dataset, model = parts

    # Load OmegaConf
    cfg_path = os.path.join(config_dir, file)
    cfg = OmegaConf.load(cfg_path)

    # Extract parameters
    entry = {}
    for key in keys_to_extract:
        val = OmegaConf.select(cfg, key)
        if val is not None:
            entry[key] = val

    # Populate nested dict
    if dataset not in standard_config:
        standard_config[dataset] = {}
    standard_config[dataset][model] = entry

# correct epoch
# standard_config["amazon_ratings"]["gcn"]["train.epoch"] = 1000

epoch_pattern = re.compile(r"\[Best Model\] Epoch: (\d+)")

# Process each file in the directory
# log_dir = "/home/yuxin/Projects/gnn-moe/processed/trial_log_epoch/result"
log_dir = "../processed/trial_log_epoch/result"

for filename in os.listdir(log_dir):
    if not filename.endswith(".txt"):
        continue
    tmp = filename.replace(".txt", "").split("_")
    if len(tmp) != 3:
        dataset, model = tmp[0], tmp[1]
    else:
        dataset, model = "_".join(tmp[:-1]), tmp[-1]
    if model not in models:
        continue
    file_path = os.path.join(log_dir, filename)

    with open(file_path, "r") as f:
        content = f.read()

    # Find all matching epoch numbers
    epochs = list(map(int, epoch_pattern.findall(content)))

    if epochs:
        avg_epoch = round(sum(epochs) / len(epochs))
        standard_config[dataset][model]["train.epoch"] = avg_epoch

# Print or save
for dataset, params in standard_config.items():
    print(f"{dataset}: {params}")
```

```python
import torch
import matplotlib.pyplot as plt
import numpy as np

selected_datasets = ["arxiv", "squirrel", "actor", "wikics",
                 "amazon_ratings", "penn94",]
# === Config ===
logit_dir = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep/logit"
fig_dir = "/home/yuxin/Projects/gnn-moe/processed/trial_sweep/figs"
trend_file = f"{logit_dir}/trend_value.pt"

for dataset in datasets:
    for sweep_key in ["model.dropout", "model.hidden_dim", "train.epoch"]:

        # === Load saved trend values ===
        trend_value = torch.load(trend_file)

        if dataset not in trend_value or sweep_key not in trend_value[dataset]:
            continue

        model_data = trend_value[dataset][sweep_key]

        plt.figure()
        for model in sorted(model_data.keys()):  # e.g., GCN, GAT
            std_val = standard_config.get(dataset, {}).get(model.lower(), {}).get(sweep_key, None)
            x_vals, y_vals = model_data[model]
            x_vals_np = np.array(x_vals)
            y_vals_np = np.array(y_vals)

            # Plot main trend
            plt.plot(x_vals_np, y_vals_np, marker='o', label=model)

            # Plot marker for standard value if defined
            if std_val is not None:
                # Check if std_val is in x_vals directly
                if std_val in x_vals_np:
                    idx = np.where(x_vals_np == std_val)[0][0]
                    std_x = std_val
                    std_y = y_vals_np[idx]
                else:
                    # Interpolate y at std_val
                    if std_val < x_vals_np.min() or std_val > x_vals_np.max():
                        continue  # Skip if std_val is outside range
                    std_y = np.interp(std_val, x_vals_np, y_vals_np)
                    std_x = std_val

                plt.plot(std_x, std_y, marker='s', markersize=8, linestyle='None',
                    markerfacecolor='red', markeredgecolor='red')

        plt.title(f"{dataset.upper()}: Error Inconsistency vs {sweep_key}")
        plt.xlabel(sweep_key)
        plt.ylabel("Error Inconsistency")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.savefig(f"{fig_dir}/{dataset}_{sweep_key.replace('.', '_')}_trend.png")
```
